```python
import requests
import json

# Azure Endpoint
url = ""  #rest end point

# API Key
api_key = ""   #primary key

# Headers
headers = {
    "Content-Type": "application/json",
    "Authorization": f"Bearer {api_key}"
}

# Input values
payload = {
    "input_data": {
        "columns": [
            "age",
            "sex",
            "bmi",
            "children",
            "smoker",
            "Claim_Amount",
            "past_consultations",
            "num_of_steps",
            "Hospital_expenditure",
            "NUmber_of_past_hospitalizations",
            "Anual_Salary",
            "region"
        ],
        "index": [0],
        "data": [
            [
                26,             # age
                "female",         # sex
                28.5,           # bmi
                1,              # children
                "no",           # smoker
                25000,          # Claim_Amount
                10,             # past_consultations
                810000,         # num_of_steps
                4800000,        # Hospital_expenditure
                0,              # NUmber_of_past_hospitalizations
                4000000,        # Anual_Salary
                "southeast"     # region
            ]
        ]
    },
    "params": {}
}

# Send request
response = requests.post(url, headers=headers, json=payload)

print("Status Code:", response.status_code)
print("Response:", response.text)
```
